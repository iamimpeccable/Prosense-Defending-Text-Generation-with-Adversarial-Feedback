```python
from datasets import Dataset, load_dataset, DatasetDict
import random
import json
```

```python
from datasets import load_from_disk

# Replace this with the actual path to your extracted folder
adv_data_path = "/content/drive/MyDrive/adversarial5000/adversarial5000"

adv_dataset = load_from_disk(adv_data_path)
print(f"Adversarial samples loaded: {len(adv_dataset)}")
```

```python
clean_dataset = load_dataset("hakurei/open-instruct-v1", split="train[:50000]")
print(f"Clean samples loaded: {len(clean_dataset)}")
```

```python
# Convert both to list of dicts
clean_data = clean_dataset.to_list()
adv_data = adv_dataset.to_list()

# Combine and shuffle
merged_data = clean_data + adv_data
random.shuffle(merged_data)
print(f"Total merged samples: {len(merged_data)}")
```

```python
with open("merged_dataset.json", "w") as f:
    json.dump(merged_data, f, indent=2)

# Optional: Download
from google.colab import files
files.download("merged_dataset.json")
```
